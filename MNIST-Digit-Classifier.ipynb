```python
# Load datasets
import pandas as pd

data_path = '/Users/dean/Desktop/Data Science/Projects/MNIST-Digit-Classifier/data'
train_df = pd.read_csv(f'{data_path}/mnist_train.csv')
test_df = pd.read_csv(f'{data_path}/mnist_test.csv')

# Check the shape and sample
print("Train shape:", train_df.shape)
print("Test shape:", test_df.shape)
print(train_df.head())
```

```python
# Check the shape and basic info of the training dataset
print(train_df.shape)
print(train_df.info())
print(train_df.describe())
```

```python
# Visualize the distribution of digit labels
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='label', data=train_df)
plt.title("Label Distribution in Training Set")
plt.xlabel("Digit Label")
plt.ylabel("Count")
plt.show()
```

```python
# Display the first 10 images from the training set
import numpy as np

plt.figure(figsize=(10, 2))
for i in range(10):
    pixels = train_df.iloc[i, 1:].values.reshape(28, 28)
    label = train_df.iloc[i, 0]
    plt.subplot(1, 10, i+1)
    plt.imshow(pixels, cmap='gray')
    plt.title(label)
    plt.axis('off')
plt.suptitle("Sample Digits from Training Set")
plt.show()
```

```python
# Plot the overall distribution of pixel intensity values
pixel_values = train_df.iloc[:, 1:].values.flatten()

plt.hist(pixel_values, bins=50, color='blue', edgecolor='black')
plt.title("Pixel Intensity Distribution")
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
plt.show()
```
